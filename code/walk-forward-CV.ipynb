```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
%matplotlib inline 
```

```python
# walk forward CV, expanding windown 
fig, axs = plt.subplots(4, 5, constrained_layout = True, figsize = (20, 10))

for i in range(4): 
    for j in range(5): 
        axs[i, j].add_patch(Rectangle((0, 0), 2, 1.2, fill = False))
        axs[i, j].axis('equal')
        axs[i, j].axis('off')
        axs[i, j].text(1, 0.6, f"Fold {j+1}", size = 40, ha = "center", va = "center")

for i in range(4): 
    for j in range(i+2): 
        
        if i+1 == j: 
            axs[i, j].add_patch(Rectangle((0, 0), 2, 1.2, fill = True, color = 'blue', alpha = 0.3))
            axs[i, j].axis('equal')
            axs[i, j].axis('off')
            # axs[i, j].text(1, 0.6, "Test", size = 40, ha = "center", va = "center")
        else: 
            axs[i, j].add_patch(Rectangle((0, 0), 2, 1.2, fill = True, color = 'green', alpha = 0.3))
            axs[i, j].axis('equal')
            axs[i, j].axis('off')
            # axs[i, j].text(1, 0.6, "Train", size = 40, ha = "center", va = "center")

plt.savefig('../plots/walk_forward_cv_expand.pdf')
```
