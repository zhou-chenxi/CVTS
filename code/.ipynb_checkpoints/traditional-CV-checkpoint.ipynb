```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
%matplotlib inline 
```

```python
fig, axs = plt.subplots(5, 6, constrained_layout = True, figsize = (40, 20))

for i in range(5): 
    for j in range(6): 
        
        if j == 0: 
            axs[i, j].add_patch(Rectangle((0, 0), 2, 1.2, fill = False, ec = None))
            axs[i, j].axis('equal')
            axs[i, j].axis('off')
            axs[i, j].text(1, 0.6, f"Fold {i+1}", size = 40, ha = "center", va = "center")
            
        else: 

            if (i+1) == j: 
                axs[i, j].add_patch(Rectangle((0, 0), 2, 1.2, fill = True, color = 'blue', alpha = 0.3))
                axs[i, j].axis('equal')
                axs[i, j].axis('off')
                axs[i, j].text(1, 0.6, "Test", size = 40, ha = "center", va = "center")
            else: 
                axs[i, j].add_patch(Rectangle((0, 0), 2, 1.2, fill = True, color = 'green', alpha = 0.3))
                axs[i, j].axis('equal')
                axs[i, j].axis('off')
                axs[i, j].text(1, 0.6, "Train", size = 40, ha = "center", va = "center")

plt.savefig('../plots/traditional_cv_horizontal.pdf')
```
