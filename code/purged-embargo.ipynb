```python
import os 
import numpy as np
import pandas as pd
import statsmodels.api as sm

import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
np.random.seed(1)
```

```python
n_samples = 365
arparams = np.array([.75, -.25])
maparams = np.array([.65, .35])
ar = np.r_[1, -arparams] # add zero-lag and negate
ma = np.r_[1, maparams] # add zero-lag
arma_process = sm.tsa.ArmaProcess(ar, ma)
y = arma_process.generate_sample(n_samples)
```

```python
df = pd.DataFrame({'y': y}, index = pd.period_range("1/1/2020", freq = "D", periods = n_samples))
```

```python
# pure purged cv 
fs = 15
ax = df.plot(linewidth = 2.5, figsize = (20, 10))
ax.set_xlabel('time', fontsize = fs)
ax.set_ylabel('y', fontsize = fs)
ax.tick_params('both', labelsize = fs)
#ax.vlines(x = '2020-07-01', ymin = -10, ymax = 10, linestyle = '--', linewidth = 2, color = 'red')
ax.vlines(x = '2020-07-10', ymin = -10, ymax = 10, linestyle = '--', linewidth = 2, color = 'red')
# ax.axvspan('2020-07-01', '2020-07-11', facecolor = 'red', alpha = 0.3)
ax.axvspan('2020-01-01', '2020-07-11', facecolor = 'green', alpha = 0.3)
ax.axvspan('2020-07-11', '2020-12-31', facecolor = 'blue', alpha = 0.3)
ax.text(x = '2020-01-03', y = -6.5, s = 'training set', fontsize = fs)
ax.text(x = '2020-07-13', y = -6.5, s = 'test set', fontsize = fs)
# ax.text(x = '2020-07-13', y = 6.2, s = 'embargo period', fontsize = fs)

# ax.annotate('purged period', xy = ('2020-07-05', 6.0), xytext = ('2020-07-13', 6.0),
#             arrowprops = dict(facecolor = 'black', arrowstyle = '->'), fontsize = fs)

ax.set_ylim((-6.9, 6.5))
ax.legend([])
plt.savefig('../plots/pure_purged_cv_before.pdf')
```
